## Questão 2
Desenvolva um sistema de banco em python. Você deve exibir 6 menus:
1. Criar nova conta. Uma nova conta deve conter ID, NOME e SALDO
2. Adicionar dinheiro a conta. Adiciona uma quantia (dita pelo usuário) a conta de um ID especifico, mostra quando ficou o saldo final.
3. Remover dinheiro da conta. Remove uma quantia (dita pelo usuário) a conta de um ID especifico, mostra quanto ficou o saldo final. Emitir um alerta caso o saldo tenha ficado negativo.
4. Mostrar saldo de um nome dado
5. Apagar uma conta (apaga todos os dados de um determinado ID)
6. Apagar todas as contas

ps: lembre-se de tratar exceções, caso o menu selecionado, ID ou nome sugerido não exista imprimir uma alerta avisando que não existe.

```
class Conta:

    def __init__(self, id, nome, saldo):
        self._nome = nome
        self._saldo = saldo
        self._id = id

    def getSaldo(self):
        return self._saldo

    def getID(self):
        return self._id

    def getNome(self):
        return self._nome

    def add_saldo(self, deposito):
        self._saldo += deposito

    def sacar(self, valor):
        self._saldo -= valor


class Banco:

    def __init__(self):
        self._contas = {}

    def buscar_conta_id(self, id):
        return self._contas.get(id)

    def buscar_conta_nome(self, nome):
        for c in self._contas:
            if self._contas[c].getNome().lower() == nome.lower():
                return self._contas[c]
        return None


    def criar_conta(self, conta:Conta):
        if self.buscar_conta_id(conta.getID()) == None:
            self._contas[conta.getID()] = conta
            print("Conta cadastrada com sucesso!!!")
        else:
            print("Já existe uma conta cadastrada com esse ID {0}".format(conta.getID()))

    def remover_conta(self,id):
        if self.buscar_conta_id(id) != None:
            del self._contas[id]
            print("A conta de ID {0} foi removida com sucesso!!!".format(id))
        else:
            print("Não foi encontrado nenhuma conta com esse ID {0}".format(id))

    def remover_todas_contas(self):
        self._contas.clear()
        print("Todas as contas foram removidas do Banco!!!")

    def depositar(self, id_conta, valor):
        conta = self.buscar_conta_id(id_conta)
        if conta != None:
            conta.add_saldo(valor)
            print("{0},deposito efetuado com sucesso, seu saldo atual é de R$ {1} reais".format(conta.getNome(),conta.getSaldo()))
        else:
            print("A conta de ID {0} não foi encontrada!!!".format(id_conta))

    def sacar(self, id_conta, valor):
        conta = self.buscar_conta_id(id_conta)
        if conta != None:
            conta.sacar(valor)
            saldo_atual = conta.getSaldo()
            print("{0}, Você sacou R$ {1} reais, seu saldo atual é de R$ {2} reais".format(conta.getNome(), valor, saldo_atual))
            if saldo_atual < 0 : print("AVISO!!!\n {0}, seu saldo atual está negativo!!!".format(conta.getNome()))
        else:
            print("A conta de ID {0} não foi encontrada!!!".format(id_conta))

    def get_saldo(self,nome_conta):
        conta = self.buscar_conta_nome(nome_conta)
        if conta != None:
            print("{0}, Seu saldo atual é de R$ {1} reais.".format(conta.getNome(), conta.getSaldo()))
        else:
            print("A conta de {0} não foi encontrada!!!".format(nome_conta))



print("Criação de objetos para TESTE")
banco = Banco()

c1 = Conta(1,"Matheus",5000)
c2 = Conta(2,"Marcos",100)
c3 = Conta(3,"Lucas",1500)

print("")
banco.criar_conta(c1)
banco.criar_conta(c2)
banco.criar_conta(c3)

print("")
banco.get_saldo("Matheus")
banco.get_saldo("João")
banco.sacar(3,300)
banco.sacar(10,300)
banco.get_saldo("Lucas")
banco.depositar(2,2000)
banco.depositar(10,2000)
banco.get_saldo("Marcos")
banco.sacar(3,1900)
banco.sacar(1,5000)
banco.remover_conta(1)
banco.remover_conta(10)
banco.get_saldo("Matheus")
banco.remover_todas_contas()
banco.get_saldo("Matheus")
banco.get_saldo("Marcos")
banco.get_saldo("Lucas")

print("")
print("")
print("-BEM VINDO AO BANCO-".center(40,"$"))
opcoes = [
    "1 - Criar nova conta",
    "2 - Depositar",
    "3 - Sacar",
    "4 - Saldo",
    "5 - Apagar uma conta",
    "6 - Apagar todas as contas",
    "0 - SAIR"]

def imprimir_opcoes():
    for op in opcoes:
        print(op)

opcao = -1
while opcao != 0:
    imprimir_opcoes()
    try:
        opcao = int(input("Digite a opção desejada: "))
        if opcao == 1:
            id = int(input("Digite o ID da conta: "))
            nome = input("Digite o nome do proprietário: ")
            saldo = float(input("Valor a ser depositado: "))
            banco.criar_conta(Conta(id, nome, saldo))
        elif opcao == 2:
            id = int(input("Digite o ID da conta: "))
            saldo = float(input("Valor a ser depositado: "))
            banco.depositar(id, saldo)
        elif opcao == 3:
            id = int(input("Digite o ID da conta: "))
            valor = float(input("Valor a ser sacado: "))
            banco.sacar(id, valor)
        elif opcao == 4:
            nome = input("Digite o nome do proprietário: ")
            banco.get_saldo(nome)
        elif opcao == 5:
            id = int(input("Digite o ID da conta: "))
            banco.remover_conta(id)
        elif opcao == 6:
            confirmacao = input("Deseja remover todas as contas (Y/N)? ")
            if confirmacao.lower() == "y":
                banco.remover_todas_contas()
        elif opcao == 0:
            print("BYE")
        else:
            print("ERroR.: Opção Inválida!!")
    except:
        print("Tipo de entrada invalida")
```
