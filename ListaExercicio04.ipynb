**1. Escreva uma função que conta a quantidade de vogais em um texto e armazena
tal quantidade em um dicionário, onde a chave é a vogal considerada.**



```
vogais = "aeiou"
def eh_vogal(letra):
    return True if letra.lower() in vogais else False

def conta_vogal(texto):
    texto = texto.replace(" ","")
    dicionario = {v:0 for v in vogais}
    for letra in texto:
        if eh_vogal(letra):
            dicionario[letra.lower()] = dicionario[letra.lower()] + 1
    return dicionario

def imp_quantidade(dicionario):
    print("Quantidade por vogal: ")
    for vogal in vogais : print("Vogal[",vogal,"]:",dicionario[vogal])

texto_teste="Lorem ipsum dolor sit amet, consectetur adipiscing elit, sed do eiusmod tempor incididunt ut labore et dolore magna aliqua. Ut enim ad minim veniam, quis nostrud exercitation ullamco laboris nisi ut aliquip ex ea commodo consequat. Duis aute irure dolor in reprehenderit in voluptate velit esse cillum dolore eu fugiat nulla pariatur. Excepteur sint occaecat cupidatat non proident, sunt in culpa qui officia deserunt mollit anim id est laborum."
print("Texto Exemplo:\n",texto_teste)
dicionario_quantidade_teste = conta_vogal(texto_teste)
print(imp_quantidade(dicionario_quantidade_teste))

texto = input("Digite um texto: ")
dicionario_quantidade = conta_vogal(texto)
print(imp_quantidade(dicionario_quantidade))
```

**2. Uma pista de Kart permite 10 voltas para cada um de 6 corredores. Escreva um programa que leia todos os tempos em segundos e os guarde em um dicionário,
onde a chave é o nome do corredor. Ao final diga de quem foi a melhor volta da

campeão é o que tem a menor média de tempos.**

```
import random

def gerar_tempo_corredores():
    dicionario = {}
    for i in range(6):
        dicionario["corredor{0}".format(i+1)] = []
        for t in range(10):
            dicionario["corredor{0}".format(i+1)].append(random.randint(50, 70))
    return dicionario

def imprimir_formatado(dicionario):
    print("Tempo de volta de cada corredor:")
    for corredor in dicionario:
        print(corredor,":",dicionario[corredor])

def calc_media(numeros):
    media = 0
    for num in numeros:
        media = media + num
    return media / len(numeros)

def melhor_volta(tempo_competidor):
    menor = tempo_competidor[0]
    volta = 0
    for i in range(len(tempo_competidor)):
        if tempo_competidor[i] <= menor:
            menor = tempo_competidor[i]
            volta = i + 1
    return {'volta':volta, 'menor_tempo':menor}

def descobrir_melhor_volta(tempo_competidores):
    melhor_competidor = list(dicionario_teste.keys()).__getitem__(0)
    menor_tempo = melhor_volta(tempo_competidores[melhor_competidor])
    for competidor in tempo_competidores:
        volta = melhor_volta(tempo_competidores[competidor])
        if volta['menor_tempo'] < menor_tempo['menor_tempo']:
            melhor_competidor = competidor
            menor_tempo = volta
        elif volta['menor_tempo'] == menor_tempo['menor_tempo']:
            #Se dois competidores tiverem o tempo de volta minima iguais, o competidor é selecionado em ordem alfabética
            if sorted([melhor_competidor,competidor])[0] != melhor_competidor:
                melhor_competidor = competidor
    return {'nome':melhor_competidor,'menor_tempo':menor_tempo['menor_tempo'],'volta':menor_tempo['volta']}


def ordenar_corredores(corredores):
    lista = []
    for c in corredores:
        if len(lista) == 0:
            lista.append((c[0],c[1]))
        else:
            media = c[1]
            ehMenor = False
            for i in range(len(lista)):
                if media < lista[i][1]:
                    lista.insert(i,(c[0], media))
                    ehMenor = True
                    break
            if not ehMenor:
                lista.append((c[0],media))
    return lista

def rakear_corredores(corredores):
    ranking = []
    for corredor in corredores:
        ranking.append((corredor,calc_media(corredores[corredor])))
    return ordenar_corredores(ranking)

def imprimir_ranking(ranking):
    for r in ranking:
        print("Corredor:",r[0], " || Média das voltas:",r[1])


print("TESTE COM CORREDORES E TEMPOS ESTATICOS")
dicionario_teste ={}
dicionario_teste["corredor1"] = [64,45,75,43,44,54,55,58,57,54]
dicionario_teste["corredor2"] = [63,42,55,43,44,54,55,58,57,54]
dicionario_teste["corredor3"] = [62,41,49,43,39,57,54,58,57,54]
dicionario_teste["corredor4"] = [65,48,57,43,44,43,44,54,57,54]
dicionario_teste["corredor5"] = [66,49,72,43,44,54,57,41,49,43]
dicionario_teste["corredor6"] = [69,50,71,43,44,54,41,49,43,54]

print(imprimir_formatado(dicionario_teste))
print(" RESUMO DA COMPETIÇÃO ".center(50,"#"))
print("Corredor com melhor voltar:",descobrir_melhor_volta(dicionario_teste))
ranking = rakear_corredores(dicionario_teste)
print("Ranking dos Competidores:")
imprimir_ranking(ranking)

print("\n\n")
print("TESTE COM CORREDORES E TEMPOS GERADOS AUTOMATICAMENTE")
corredores_gerados = gerar_tempo_corredores()
print(imprimir_formatado(corredores_gerados))
print(" RESUMO DA COMPETIÇÃO ".center(50,"#"))
print("Corredor com melhor voltar:",descobrir_melhor_volta(corredores_gerados))
ranking = rakear_corredores(corredores_gerados)
print("Ranking dos Competidores:")
imprimir_ranking(ranking)
```

**3. Escreva um programa para armazenar uma agenda de telefones em um dicionário. Cada pessoa pode ter um ou mais telefones e a chave do dicionário é o nome da pessoa.**

```
agenda = {
    'alice': {'Alice' : ['(84)99999-9901','(84)99999-9902']},
    'miguel': {'Miguel'  : ['(84)99999-9911']},
 	'sophia': {'Sophia'  : ['(84)99999-9921']},
    'arthur': {'Arthur' : ['(84)99999-9931']},
    'helena': {'Helena' : ['(84)99999-9941']},
    'bernardo': {'Bernardo' : ['(84)99999-9951']},
    'valentina': {'Valentina' : ['(84)99999-9961']},
    'heitor': {'Heitor' : ['(84)99999-9971']},
    'joãocarlos': {'João Carlos' : ['(84)99999-9981']},
}

key = lambda nome : nome.lower().replace(" ","")

def isCadastrado(nome):
    return not agenda.get(key(nome), False) == False

def imprimir_contato(contato):
    nome = list(contato.keys())[0]
    print("Nome:",nome)
    for c in contato[nome]:
        print(" - ",c)

def imprimir_contatos():
    for c in agenda:
        imprimir_contato(agenda[c])


def cadastrar(nome,numero):
    chave = key(nome)
    if isCadastrado(chave):
        agenda[chave][nome].append(numero)
    else:
        agenda[chave] = {nome: [numero]}

def consultar(nome):
    chave = key(nome)
    if isCadastrado(chave):
        print("==================")
        return imprimir_contato(agenda[chave])
    else:
        return "Não foi encontrado nenhum contato com esse nome!!!"

print("LISTA DE CONTATOS".center(40,'-'))
imprimir_contatos()

contato = input("Digite o nome do novo contato: ")
numero = ""
while numero != "-1":
    numero = input("Digite um número para o contato ou -1 para sair: ")
    if numero != "-1" :
        cadastrar(contato,numero)
        print("Cadastro Efetuado com Sucesso!!!")

print("LISTA DE CONTATOS".center(40,'-'))
imprimir_contatos()
contato = input("Buscar contato: ")
print(consultar(contato))
```

**4. Escreva um programa para armazenar em um dicionário o nome do aluno como
chave e seu cpf como valor. Faça três funções: 1. Que retorna o cpf de um
determinado aluno , 2. Que adiciona um novo aluno e cpf 3. Que imprima o
dicionário todo.**

```
alunos = {
    'alice': {'nome':'Alice', 'cpf': '111.111.111-12'},
    'miguel': {'nome':'Miguel', 'cpf': '111.111.111-13'},
 	'sophia': {'nome':'Sophia', 'cpf': '111.111.111-14'},
    'arthur': {'nome': 'Arthur', 'cpf': '111.111.111-15'},
    'helena': {'nome': 'Helena', 'cpf': '111.111.111-16'},
    'bernardo': {'nome': 'Bernardo', 'cpf': '111.111.111-17'},
    'valentina': {'nome': 'Valentina', 'cpf': '111.111.111-18'},
    'heitor': {'nome': 'Heitor', 'cpf': '111.111.111-19'},
    'joãocarlos': {'nome': 'João Carlos', 'cpf': '111.111.111-20'},
}
key = lambda nome : nome.lower().replace(" ","")

def isCadastrado(nome):
    return not alunos.get(key(nome), False) == False


def imprimir_alunos():
    for a in alunos:
        print("Nome:",alunos[a]['nome']," :: CPF:",alunos[a]['cpf'])

def add_aluno(nome,cpf):
    chave = key(nome)
    if not isCadastrado(chave):
        alunos[chave] = {'nome':nome,'cpf':cpf}
        return True
    else:
        return False

def cpf_aluno(nome):
    chave = key(nome)
    if isCadastrado(chave):
        return alunos[chave]['cpf']
    else:
        return None

print("LISTA DE ALUNOS".center(40,'-'))
imprimir_alunos()

aluno = input("Digite o nome do novo aluno: ")
cpf = input("Agora, digite o CPF: ")
if add_aluno(aluno,cpf):
    print("Cadastro Efetuado com Sucesso!!!")
else:
    print("Não foi possível efetuar o cadastro, Já existe um aluno cadastrado com esse nome!!!")

print("LISTA DE ALUNOS".center(40,'-'))
imprimir_alunos()
aluno = input("Buscar aluno: ")
cpf = cpf_aluno(aluno)
print("CPF: {0}".format(cpf) if cpf != None else "Não foi encontrado nenhum aluno com esse nome!!!")
```

**5. Escreva um programa que recebe um dicionário com a chave nome e o valor idade e imprima uma lista dos nomes por ordem (crescente) de idade.**

```
pessoas = {
    'Alice' : 48,
    'Miguel' : 16,
 	'Sophia' : 28,
    'Arthur' : 12,
    'Helena' : 37,
    'Bernardo' : 43,
    'Valentina' : 58,
    'Heitor' : 63,
    'João Carlos' : 33,
}

def ordenar_por_idade(pessoas):
    lista = []
    for p in pessoas:
        if len(lista) == 0:
            lista.append((p,pessoas[p]))
        else:
            idade = pessoas[p]
            ehMenor = False
            for i in range(len(lista)):
                if idade < lista[i][1]:
                    lista.insert(i,(p, idade))
                    ehMenor = True
                    break
            if not ehMenor:
                lista.append((p, idade))
    return lista

ordenado  = ordenar_por_idade(pessoas)
for ord in ordenado:
    print(ord[0],ord[1],"anos")
```
