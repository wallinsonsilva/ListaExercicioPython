## Questão 1:

Desenvolva um jogo da forca. O programa terá uma lista de palavras lidas de uma lista e escolherá uma palavra aleatoriamente. O jogador vai tentar descobrir essa palavra, mas só poderá errar 6 vezes antes de ser enforcado.

Digite uma letra: A -> Você errou pela 1ª vez. Tente de novo!

Digite uma letra: O A palavra é: O

Digite uma letra: E A palavra é: E _ O

Digite uma letra: S -> Você errou pela 2ª vez. Tente de novo!

Quando errar pela 6ª vez imprimir: Enforcado e exibir a palavra correta


```
import random
estagios = ['''
    +------+
    |      |
    |
    |
    |
    |
=========''',
'''
    +------+
    |      |
    |      0
    |
    |
    |
=========''',
'''
    +------+
    |      |
    |      0
    |      |
    |
    |
=========''',
'''
    +------+
    |      |
    |      0
    |     /|
    |
    |
=========''',
'''
    +------+
    |      |
    |      0
    |     /|\\
    |
    |
=========''',
'''
    +------+
    |      |
    |      0
    |     /|\\
    |     /
    |
=========''',
'''
    +------+
    |      |
    |      0
    |     /|\\
    |     / \\
    |
=========
''']

palavras = [
"Acender","Afilhado","Ardiloso","Aspero","Assombraçao","Asterisco","Basquete",
    "Caminho","Champanhe","Chiclete","Chuveiro","Coelho","Contexto","Convivencia",
    "Coraçao","Desalmado","Eloquente","Esfirra","Esquerdo","Exceçao","Fugaz","Gororoba"
]

def selecionar_palavra():
    return palavras[random.randint(0, len(palavras)-1)]

def criar_mascara(palavra):
    mascara = []
    for letra in palavra:
        mascara.append("_")
    return mascara

def imprimir_forca(quant_erros):
    print(estagios[quant_erros])

def letra_in_palavra(chute, palavra):
    if chute.lower() in palavra.lower():
        return True
    else:
        return False

def preencher_palavra_secreta(chute,palavra,mascara):
    cont = 0
    for letra in palavra:
        if chute.lower() == letra.lower():
            mascara[cont] = chute.upper()
        cont += 1
    return mascara




palavra = selecionar_palavra()
mascara = criar_mascara(palavra)
chances = 6
quantidade_erros = 0
acertos = 0
jogada = 0
lista_chutes = []

print(" JOGO DA FORCA ".center(40,"#"))
imprimir_forca(quantidade_erros)
while  "_" in mascara and quantidade_erros < chances:
    print("Palavra Secreta:",mascara)
    chute = input("Digite uma letra: ")
    if not chute.lower() in "".join(lista_chutes).lower():
        if letra_in_palavra(chute, palavra):
            mascara = preencher_palavra_secreta(chute, palavra, mascara)
            acertos += 1
            print("Você acertou pela {0}ª vez!!!".format(acertos))
        else:
            quantidade_erros += 1
            print("Você errou pela {0}ª vez!!!".format(quantidade_erros))
            imprimir_forca(quantidade_erros)
        lista_chutes.append(chute)
    else:
        print("Você ja chutou a letra {0}".format(chute.upper()))
    print("Letras digitadas: ",lista_chutes)




if not "_" in mascara:
    print("Você conseguiu descobrir que a palavra secreta era '{0}'".format("".join(palavra)))
else:
    print("Enforcado, a palavra era '{0}'".format("".join(palavra)))
    imprimir_forca(quantidade_erros)
    print("Seus acertos:",mascara)
```
